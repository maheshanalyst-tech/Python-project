3. Exploratory Data Analysis (EDA) and Visualization

Campaign Performance:

➢ Plot a scatter plot to visualize the relationship between Acquisition_Cost and ROI.


```python

plt.figure(figsize=(12,8))

sns.scatterplot(data=df, x='Acquisition_Cost', y='ROI', hue='Campaign_Type')

plt.title('Acquisition Cost vs ROI by Campaign Type')

plt.show()
```

➢ Create a bar chart to visualize the average Conversion_Rate for different
Channel_Used, categorized by Campaign_Type.


```python
plt.figure(figsize=(14,8))

sns.barplot(data=df, x='Channel_Used', y='Conversion_Rate', hue='Campaign_Type')

plt.title('Average Conversion Rate by Channel and Campaign Type')

plt.xticks(rotation=45)

plt.show()
```

➢ Visualize the distribution of Engagement_Score across different Campaign_Type
using a box plot.

```python
plt.figure(figsize=(11,7))

sns.boxplot(data=df, x='Campaign_Type', y='Engagement_Score')

plt.title('Engagement Score Distribution by Campaign Type')

plt.show()
```

➢ Analyze the average ROI by Company using a bar chart to compare the profitability of campaigns conducted by different companies.


```python
plt.figure(figsize=(10,8))

df.groupby('Company')['ROI'].mean().sort_values().plot(kind='bar')

plt.title('Average ROI by Company')

plt.ylabel('Average ROI')

plt.show()
```

➢ Examine the correlation between Engagement_Score and Conversion_Rate
using a heatmap.


```python
plt.figure(figsize=(12,8))

sns.heatmap(df[['Engagement_Score', 'Conversion_Rate']].corr(), annot=True, cmap='coolwarm')

plt.title('Correlation between Engagement Score and Conversion Rate')

plt.show()
```

Customer Segmentation:

➢ Create a count plot to visualize the distribution of Target_Audience.

```python
plt.figure(figsize=(10,8))

sns.countplot(data=df, y='Target_Audience')

plt.title('Distribution of Target Audience')

plt.show()
```

 ➢ Identify which Customer_Segment has the highest Conversion_Rate for each Language using a bar chart.

```python
plt.figure(figsize=(14,7))

sns.barplot(data=df, x='Language', y='Conversion_Rate', hue='Customer_Segment')

plt.title('Conversion Rate by Language and Customer Segment')

plt.show()
```

➢ Visualize the distribution of Acquisition_Cost across each Customer_Segment,
categorized by Channel_Used, using a box plot.


```python
plt.figure(figsize=(13,9))

sns.boxplot(data=df, x='Customer_Segment', y='Acquisition_Cost', hue='Channel_Used')

plt.title('Acquisition Cost Distribution by Customer Segment and Channel')

plt.xticks(rotation=90)

plt.legend(bbox_to_anchor=(1.09, 1), loc='upper right')

plt.show()
```

➢ Analyze average Conversion_Rate by Language using a bar chart to compare the effectiveness of campaigns conducted in different languages.

```python
plt.figure(figsize=(12,8))

df.groupby('Language')['Conversion_Rate'].mean().sort_values().plot(kind='bar')

plt.title('Average Conversion Rate by Language')

plt.ylabel('Average Conversion Rate')

plt.show()
```

Channel Effectiveness:

➢ Compare the Engagement_Score for different Channels_Used, segmented by
Campaign_Type, using a bar chart.


```python
plt.figure(figsize=(12,10))

sns.barplot(data=df, x='Channel_Used', y='Engagement_Score', hue='Campaign_Type')

plt.title('Engagement Score by Channel and Campaign Type')

plt.xticks(rotation=45)

plt.show()
```

➢ Show the distribution of total ROI across different Channels_Used using a pie
chart.


```python
plt.figure(figsize=(12,12))

df.groupby('Channel_Used')['ROI'].sum().plot(kind='pie', autopct='%1.1f%%')

plt.title('Total ROI Distribution by Channel')

plt.ylabel('')

plt.show()
```

➢ Plot a scatter plot to show the relationship between Clicks and Impressions
for each Channel_Used.

```python
plt.figure(figsize=(10,6))

sns.scatterplot(data=df, x='Clicks', y='Impressions', hue='Channel_Used', size='ROI')

plt.title('Clicks vs Impressions by Channel (Size represents ROI)')

plt.show()
```

Time-Based Analysis:

➢ Plot the distribution of Duration using a histogram.

```python
plt.figure(figsize=(10,12))\

sns.histplot(data=df, x='Duration', bins=15)

plt.title('Distribution of Campaign Duration')

plt.show()
```

➢ Analyze how the overall Conversion_Rate has changed over Date for each
Company using a line chart.


```python
plt.figure(figsize=(15,10))

for company in df['Company'].unique():

    company_data = df[df['Company'] == company]

    company_data = company_data.groupby('Date')['Conversion_Rate'].mean().reset_index()

    sns.lineplot(data=company_data, x='Date', y='Conversion_Rate', label=company)

plt.title('Conversion Rate Trend Over Time by Company')

plt.legend(bbox_to_anchor=(1.05, 1), loc='lower right')

plt.show()
```

➢ Examine the trend of Engagement_Score over Date with a line chart.

```python
plt.figure(figsize=(15,10))

df.groupby('Date')['Engagement_Score'].mean().plot()

plt.title('Engagement Score Trend Over Time')

plt.ylabel('Average Engagement Score')

plt.show()
```

Geographic Analysis:

➢ Determine which location has the highest Acquisition_Cost using a bar chart.


```python
plt.figure(figsize=(12,6))

df.groupby('Location')['Acquisition_Cost'].mean().sort_values(ascending=False).plot(kind='bar')

plt.title('Average Acquisition Cost by Location')

plt.ylabel('Average Acquisition Cost')

plt.show()
```

➢ Visualize the Conversion_Rate by different Location, categorized by Target_Audience, using a bar chart.


```python
plt.figure(figsize=(12,6))

sns.barplot(data=df, x='Location', y='Conversion_Rate', hue='Target_Audience')

plt.title('Conversion Rate by Location and Target Audience')

plt.xticks(rotation=45)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

➢ Illustrate the proportion of ROI by Location using a pie chart.

```python
plt.figure(figsize=(9,9))

df.groupby('Location')['ROI'].sum().plot(kind='pie', autopct='%1.1f%%')

plt.title('Total ROI Distribution by Location')

plt.ylabel('')

plt.show()
```
